```python
# Detecting outliers using IQR
def detect_outliers_iqr(dataa, column):
    Q1 = dataa[column].quantile(0.25)
    Q3 = dataa[column].quantile(0.75)
    IQR = Q3 - Q1

    # Define lower and upper limits
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    # Filter outliers
    outliers = dataa[(dataa[column] < lower_limit) | (dataa[column] > upper_limit)]
    
    print(f"{column}: {len(outliers)} outliers found")
    return outliers

# Apply IQR method on selected columns
for col in dataa.select_dtypes(include='number').columns:
    detect_outliers_iqr(dataa, col)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from sklearn.linear_model import LogisticRegression
```

```python
X = dataa.drop('status', axis=1)
y = dataa['status']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.3, random_state=45)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
smote = SMOTE(random_state=42)
```

```python
X_train_smote, y_train_smote = smote.fit_resample(X_train_scaled, y_train)
```

```python
model = LogisticRegression()
model.fit(X_train_scaled,y_train)
```

```python
y_pred = model.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
accuracy = accuracy_score(y_test, y_pred)
```

```python
report = classification_report(y_test,y_pred)
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
print(f'Accuracy :' , accuracy)
print(f'classification_report: ', report)
print(f'confusion_matrix :', cm)
```

```python
plt.figure(figsize = (8,6))
plt.hist([y_test, y_pred], label=['Actual, predicted'], bins = 10, alpha = 0.7)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python
report = classification_report(y_test, y_pred, output_dict=True)
report_df = pd.DataFrame(report).T

# Visualizing the classification report
plt.figure(figsize=(8, 5))
sns.heatmap(report_df.iloc[:-1, :].T, annot=True, cmap='coolwarm', cbar=False)
plt.title('Classification Report Visualization')
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc

# Predict probabilities
y_prob = model.predict_proba(X_test)[:, 1]

# ROC Curve
fpr, tpr, _ = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(7, 5))
plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend()
plt.grid(True)
plt.show()
```
